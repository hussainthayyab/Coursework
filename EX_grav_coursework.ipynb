```python
from pylab import*
import cmath
import math
import numpy as np
```

```python
def resp(f):
  l = 4*(10**3)
  c = 3*(10**8)
  lamda = 1064 * (10**-9)
  pi = math.pi
  return (4*l*math.pi)*(np.sinc(2*(l/c)*f))*cmath.exp(2*pi*1j*(l/c)*f)

f = logspace(1,6,10000) # frequencies from 10**1 to 10**5
f2 = np.vectorize(resp)
plot(f,log10(f2(f)))

plt.xlabel('frequency')
plt.ylabel('phase difference/ h (rad)')
plt.xscale('log')
plt.yscale('log')

#focus on frequency range 10^4 to 10^5
```
